$ax^2 + bx + c = 0$

$x^2 + 2x + 1 = 0$

# pq-formeln

$x = -\frac{p}{2} \pm \sqrt{(\frac{p}{2})^2 - q}$

```python
import numpy as np

q_term = [1, 2, 1]

def quadratic_solver(coeff):
    a, b, c = coeff

    x1 = np.divide(-b + np.sqrt(np.square(b) - 4*a*c), 2*a)
    x0 =  np.divide(-b - np.sqrt(np.square(b) - 4*a*c), 2*a)
    return x0, x1

print(quadratic_solver(q_term))
```

```python
import matplotlib.pyplot as plt

pasta_per_year = []

for year in range(140):
    pasta_comsumption = .791 * np.exp(.0526*year)
    pasta_per_year.append(pasta_comsumption)
plt.plot(range(1960, 2100), pasta_per_year)
plt.xlabel("Year")
plt.ylabel("Pasta comsumption")
plt.title("Pasta comsumption in kg per person per year from 1960")

plt.show()
```
