```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#a)
college = pd.read_csv("../../Data/Databehandling/College.csv")

college
```

```python
#b)
college2 = pd.read_csv("../../Data/Databehandling/College.csv", index_col=0)
college3 = college.rename({"Unnamed: 0": "College"}, axis = 1)
college3 = college3.set_index("College")

college = college3

college
```

```python
#c)
college.describe()
```

```python
#d)
pd.plotting.scatter_matrix(college[["Top10perc", "Apps", "Enroll"]])
plt.show()
```

```python
#e)

college.boxplot(column=["Outstate"], by="Private", figsize=(8,6))
plt.show()
```

```python
# f)
college["Elite"] = pd.cut(college["Top10perc"], [0, 50, 100], labels=["No", "Yes"])

college.value_counts(["Elite"])

college.boxplot(column=["Outstate"], by="Elite", figsize=(8,6))
plt.show()
```

```python
# g)

plt.subplot(2, 2, 1)
plt.hist(college["Apps"], bins=50)
plt.title("Apps")

plt.subplot(2, 2, 2)
plt.hist(college["Accept"], bins=75)
plt.title("Accept")

plt.subplot(2, 2, 3)
plt.hist(college["Enroll"], bins=100)
plt.title("Enroll")

plt.subplot(2, 2, 4)
plt.hist(college["Top10perc"], bins=125)
plt.title("Top10perc")

plt.tight_layout()
plt.show()
```

```python
# h)
# Elit, Personal, Expend

plt.figure(figsize=(10,12))
plt.subplot(3, 2, 1)
sns.barplot(x="Elite", y="Personal", data=college, errorbar=None)
plt.title("Jämför Personliga utgifter mellan Elite & Non-Elite")
plt.xlabel("Elite status")
plt.ylabel("Beräknade Personliga utgifter")

plt.subplot(3, 2, 2)
sns.barplot(x="Elite", y="Expend", data=college, errorbar=None)
plt.title("Jämför undervisnings utgifter mellan Elite & Non-Elite")
plt.xlabel("Elite status")
plt.ylabel("Utgifter för undervisning per elev")

plt.subplot(3, 2, 3)
sns.barplot(x="Private", y="Personal", data=college, errorbar=None)
plt.title("Jämför Personliga utgifter mellan Privat & Offentlig skola")
plt.xlabel("Private")
plt.ylabel("Beräknade Personliga utgifter")

plt.subplot(3, 2, 4)
sns.barplot(x="Private", y="Expend", data=college, errorbar=None)
plt.title("Undervisnings utgifter mellan Privat & Offentlig skola")
plt.xlabel("Private")
plt.ylabel("Utgifter för undervisning per elev")

plt.subplot(3, 2, 5)
sns.barplot(x="Private", y="perc.alumni", data=college, errorbar=None)
plt.title("Donationer till Privat och Offentlig skola")
plt.xlabel("Private")
plt.ylabel("Procent av alumni/doktorander som donerar")

plt.subplot(3, 2, 6)
sns.barplot(x="Elite", y="perc.alumni", data=college, errorbar=None)
plt.title("Donationer till Elite & Non-Elite")
plt.xlabel("Elite status")
plt.ylabel("Procent av alumni/doktorander som donerar")

plt.tight_layout()
plt.show()
```

#### Summary
Jag valde att jämföra personliga kostnader, undervisnings utgifter och donationer för de college med Elite status samt jämföra samma kostnader, utgifter och donationer mellan Privat och Offentlig college. 

#### Exploring
